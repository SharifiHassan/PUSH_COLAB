✅ How to Push a Google Colab Project to GitHub (Clean Setup)

🟩 Step 1: Create a New GitHub Repository
1. Go to 👉 https://github.com/new
2. Repository name: e.g., ML_Journey_Day3
3. Do NOT check: ○ "Initialize this repository with a README"
4. Click Create repository Copy the HTTPS URL (e.g., https://github.com/sharifihassan/PUSH_COLAB.git)

##🔐  Step 2: Create a GitHub Personal Access Token (only once)
1. Go to 👉 https://github.com/settings/tokens

2. Choose Fine-grained tokens > Generate new token

3. Set:

  *  Token name: Colab GitHub Token

  *  Repo access: "All repositories"
  
  *  Expiration: Optional

4. Under Permissions, enable:

* ✅ Contents: Read and write

5. Click Generate token and copy it immediately

````python
import getpass

token = getpass.getpass("Enter your GitHub token: ")
username = "sharifihassan"
repo = "PUSH_COLAB"

remote_url = f"https://{token}@github.com/{username}/{repo}.git"

!git remote add origin $remote_url
````

# 🧼 Common Cleanup Commands

If you mess up:

**Removes incorrect remote**

### `!git remote remove origin`
**Wipe all staged files**

### `!git reset --hard`

**Wipe untracked files/folders**

### `!git clean -fd`
